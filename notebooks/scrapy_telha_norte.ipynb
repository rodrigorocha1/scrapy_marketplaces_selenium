```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver import ActionChains
from time import sleep
from datetime import datetime
```

```python
servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
navegador.maximize_window()
```

```python
navegador.get('https://www.telhanorte.com.br/')
```

```python
busca_produto = navegador.find_element(
    By.CLASS_NAME, 'vtex-styleguide-9-x-input')

busca_produto.send_keys('Telha', Keys.ENTER)
```

```python
WebDriverWait(navegador, 10).until(EC.element_to_be_clickable(
    (By.ID, 'onetrust-accept-btn-handler'))).click()
```

```python
slider = navegador.find_element(
    By.CLASS_NAME, 'vtex-slider__selector-container')
move = webdriver.ActionChains(navegador)
move.click_and_hold(slider).move_by_offset(3, 0).release().perform()
```

- Paginação

```python
while True:
    try:
        navegador.find_element(
            By.XPATH, '/html/body/div[2]/div/div[1]/div/div[1]/div/div[4]/section/div[2]/div/div[3]/div/a/div'
        ).click()
    except Exception as E:
        break
```

```python
lista_de_produtos = navegador.find_elements(
    By.CLASS_NAME, 'vtex-search-result-3-x-galleryItem')
len(lista_de_produtos)
```

```python
nome_produtos = navegador.find_elements(
    By.CLASS_NAME, 'vtex-product-summary-2-x-productBrand')
nome_produtos[1].text
precos = navegador.find_elements(
    By.CLASS_NAME, 'telhanorte-telha-store-app-1-x-best-price-price-vitrini')
url_produtos = navegador.find_elements(
    By.CLASS_NAME, 'vtex-product-summary-2-x-clearLink'
)
url_imagens = navegador.find_elements(
    By.XPATH, '//section/a/article/div[2]/img'
)
```

```python
for produto, preco, url_produto, url_imagem in zip(nome_produtos, precos, url_produtos, url_imagens):
    print(produto.text)
    print(preco.text)
    print(url_produto.get_attribute('href'))
    print(url_imagem.get_attribute('src'))
    print(url_produto.get_attribute('href').split('-')[-1].replace('/p', ''))
    print()
```
